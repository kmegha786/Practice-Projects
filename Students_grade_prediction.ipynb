# Training our model

Separating features and label as x and y

```python
x=df.drop(columns=['CGPA'],axis=1)
y=df['CGPA']
```

# Data Standardisation

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x_scaled=scaler.fit_transform(x)
```

# Check for Multicolliernity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['vif1']=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif['features']=x.columns
vif
```

Variance Inflation Factor is less than 5 for all columns so there is no multicollinearity

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
lr=LinearRegression()
```

# Finding the best random state

```python
max_acc=0
max_rs=0

for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=i,test_size=0.20)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred_lr=lr.predict(x_test)
    accuracy=r2_score(y_test,pred_lr)
    print('Testing accuracy', accuracy,'random state',i)
    
    if accuracy>max_acc:
        max_acc=accuracy
        max_rs=i
        print('max_accuracy',max_acc,'max_random_state',i)
```

# Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```
